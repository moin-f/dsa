```python
#Q3a.2 Write Python program to evaluate interpolate value f(3) of the given data.
# X 0 1 2 5   y=f(x)  2 3 12 147

def lagrange_interpolation(x_values, y_values, x):
    n = len(x_values)
    result = 0.0
    
    for i in range(n):
        # Calculate the Lagrange basis polynomial
        term = y_values[i]
        for j in range(n):
            if j != i:
                term *= (x - x_values[j]) / (x_values[i] - x_values[j])
        result += term

    return result

# Given data points
x_values = [0, 1, 2, 5]
y_values = [2, 3, 12, 147]

# Evaluate f(3)
x_to_interpolate = 3
interpolated_value = lagrange_interpolation(x_values, y_values, x_to_interpolate)

print(f"The interpolated value f(3) is: {interpolated_value:.6f}")


```

```python
#Q3b 1 Write Python program to estimate the value of the integral 2 to 10 (1/(1 + x)) dx using Trape- zoidal rule (n = 8) .
def f(x):
    return 1 / (1 + x)

def trapezoidal_rule(a, b, n):
    h = (b - a) / n  # Calculate the width of each subinterval
    integral = (f(a) + f(b)) / 2  # Initialize with the first and last term

    # Apply the Trapezoidal rule
    for i in range(1, n):
        x = a + i * h
        integral += f(x)  # Sum the middle terms

    integral *= h  # Multiply by the width of the subintervals
    return integral

# Define the interval [a, b] and number of subintervals n
a = 2
b = 10
n = 8  # Number of subintervals

# Calculate the integral using the Trapezoidal rule
result = trapezoidal_rule(a, b, n)
print(f"The estimated value of the integral is: {result:.6f}")
```

```python
#Q3b 2.Write Python program to evaluate f(2.8) using backward difference formula of the given data.
# x  0 1 2 3   y=f(x)  1 0 1 10
def backward_difference(x_values, y_values, x):
    n = len(x_values)
    
    # Calculate the backward differences
    backward_differences = [[0] * n for _ in range(n)]
    for i in range(n):
        backward_differences[i][0] = y_values[i]

    for j in range(1, n):
        for i in range(n - j):
            backward_differences[i][j] = backward_differences[i + 1][j - 1] - backward_differences[i][j - 1]

    # Calculate the value of f(x) using the backward difference formula
    h = x_values[1] - x_values[0]  # Assuming uniform spacing
    index = n - 1  # Start with the last index
    result = backward_differences[index][0]

    # Calculate the backward difference formula
    for k in range(1, n):
        if index - k < 0:  # Check if index is within bounds
            break
        term = backward_differences[index - k][k]
        result += term * (x - x_values[index]) / h ** k
    
    return result

# Given data points
x_values = [0, 1, 2, 3]
y_values = [1, 0, 1, 10]

# Evaluate f(2.8)
x_to_evaluate = 2.8
evaluated_value = backward_difference(x_values, y_values, x_to_evaluate)

print(f"The evaluated value f(2.8) using backward difference is: {evaluated_value:.6f}")

```
