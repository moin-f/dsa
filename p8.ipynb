```python
#Q.3a.1 Write Python program to estimate the value of the integral 1 to 10 (x² + 5x)dx using Simpson's (1/3)rd rule (n=5).

def f(x):
    return x**2 + 5*x

def simpsons_rule(a, b, n):
    if n % 2 != 0:
        raise ValueError("n must be an even number for Simpson's (1/3) rule.")
    
    h = (b - a) / n
    integral = f(a) + f(b)
    
    # Apply Simpson's (1/3) rule formula
    for i in range(1, n):
        x = a + i * h
        if i % 2 == 0:
            integral += 2 * f(x)
        else:
            integral += 4 * f(x)
    
    integral *= h / 3
    return integral

# Define the interval [a, b] and number of subintervals n
a = 1
b = 10
n = 6  # Changed to an even number

# Calculate the integral using Simpson's (1/3)rd rule
result = simpsons_rule(a, b, n)
print(f"The estimated value of the integral is: {result:.6f}")
```

```python
#Q.3a2 Write Python program to evaluate interpolate value f(2.5) of the given data.
def lagrange_interpolation(x, data_points):
    """
    Evaluates the interpolated value at x using the given data points
    based on the Lagrange Interpolation formula.
    """
    result = 0
    n = len(data_points)
    
    for i in range(n):
        # Compute the Lagrange basis polynomial L_i(x)
        term = data_points[i][1]
        for j in range(n):
            if i != j:
                term *= (x - data_points[j][0]) / (data_points[i][0] - data_points[j][0])
        
        # Add the term to the result
        result += term
    
    return result

# Given data points as (x, f(x)) pairs
data_points = [(1, 2), (2, 3), (3, 5), (4, 4)]  # Replace with your specific data points

# Value to interpolate
x = 2.5

# Calculate the interpolated value at x = 2.5
interpolated_value = lagrange_interpolation(x, data_points)
print(f"The interpolated value of f(2.5) is: {interpolated_value}")

```

```python
#Q3.b1 Write Python program to obtained the approximate real root of x³ - 4x - 9= 0 by using Regula-falsi method.
def f(x):
    return x**3 - 4*x - 9

def regula_falsi(a, b, tol, max_iter):
    if f(a) * f(b) >= 0:
        print("The function has the same sign at the endpoints a and b.")
        return None

    for i in range(max_iter):
        # Calculate the false position
        c = a - (f(a) * (b - a)) / (f(b) - f(a))
        
        # Check if we found the root or if the interval is small enough
        if f(c) == 0 or abs(f(c)) < tol:
            return c

        # Update the interval
        if f(c) * f(a) < 0:
            b = c
        else:
            a = c
            
    # Return the approximation after the max iterations
    return c

# Initial guesses for the root
a = 2
b = 3

# Tolerance level and maximum iterations
tolerance = 1e-6
max_iterations = 100

# Calculate the root using Regula-Falsi method
root = regula_falsi(a, b, tolerance, max_iterations)

if root is not None:
    print(f"The approximate real root of the equation is: {root:.6f}")
else:
    print("The method did not converge.")

```

```python
#Q.3b2 Write Python program to evaluate fourth order forward difference of the given data.
# x  1 2 3 4 5  y=f(x) 40 60 65 50 18
def forward_difference_table(x_values, y_values):
    n = len(y_values)
    difference_table = [[0 for _ in range(n)] for _ in range(n)]

    # Initialize first column with y_values
    for i in range(n):
        difference_table[i][0] = y_values[i]

    # Calculate forward differences
    for j in range(1, n):
        for i in range(n - j):
            difference_table[i][j] = difference_table[i + 1][j - 1] - difference_table[i][j - 1]

    return difference_table

# Given data points
x_values = [1, 2, 3, 4, 5]
y_values = [40, 60, 65, 50, 18]

# Construct forward difference table
difference_table = forward_difference_table(x_values, y_values)

# Display the forward difference table
print("Forward Difference Table:")
for row in difference_table:
    print(row)

# Extract the fourth order forward difference
fourth_order_forward_difference = difference_table[0][4]
print(f"\nThe fourth order forward difference is: {fourth_order_forward_difference}")

```
