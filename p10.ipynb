```python
#Q.3a.1 Write Python program to estimate the value of the integral 0 to 1  (1/(1+x2)dx using Simp- son's (1/3)rd rule (n=6).
def f(x):
    return 1 / (1 + x**2)

def simpsons_rule(a, b, n):
    if n % 2 != 0:
        raise ValueError("n must be an even number for Simpson's (1/3) rule.")
    
    h = (b - a) / n
    integral = f(a) + f(b)
    
    # Apply Simpson's (1/3) rule formula
    for i in range(1, n):
        x = a + i * h
        if i % 2 == 0:
            integral += 2 * f(x)
        else:
            integral += 4 * f(x)
    
    integral *= h / 3
    return integral

# Define the interval [a, b] and number of subintervals n
a = 0
b = 1
n = 6  # Number of subintervals (must be even)

# Calculate the integral using Simpson's (1/3)rd rule
result = simpsons_rule(a, b, n)
print(f"The estimated value of the integral is: {result:.6f}")

```

```python
#Q3a.2 Write Python program to evaluate fourth order forward difference of the given data.
# x 1 2 3 4 5  y=f(x) 41 62 65 50 17 
def forward_difference_table(x_values, y_values):
    n = len(y_values)
    difference_table = [[0 for _ in range(n)] for _ in range(n)]

    # Initialize the first column with y_values
    for i in range(n):
        difference_table[i][0] = y_values[i]

    # Calculate forward differences
    for j in range(1, n):
        for i in range(n - j):
            difference_table[i][j] = difference_table[i + 1][j - 1] - difference_table[i][j - 1]

    return difference_table

# Given data points
x_values = [1, 2, 3, 4, 5]
y_values = [41, 62, 65, 50, 17]

# Construct forward difference table
difference_table = forward_difference_table(x_values, y_values)

# Display the forward difference table
print("Forward Difference Table:")
for row in difference_table:
    print(row)

# Extract the fourth order forward difference if available
if len(difference_table) > 4:
    fourth_order_forward_difference = difference_table[0][4]
    print(f"\nThe fourth order forward difference is: {fourth_order_forward_difference}")
else:
    print("\nFourth order forward difference is not available (not enough data points).")
```

```python
#Q.3b.1 Write Python program to obtained the approximate real root of x³-2x-5 = 0 in [2,3] using Regula-falsi method.

def f(x):
    return x**3 - 2*x - 5

def regula_falsi_method(a, b, tol=1e-6, max_iter=100):
    if f(a) * f(b) >= 0:
        raise ValueError("f(a) and f(b) must have opposite signs.")

    for i in range(max_iter):
        # Calculate the root using the Regula-Falsi formula
        c = b - (f(b) * (a - b)) / (f(a) - f(b))

        # Check for convergence
        if abs(f(c)) < tol:
            return c

        # Update the interval
        if f(c) * f(a) < 0:
            b = c
        else:
            a = c

    raise ValueError("Maximum iterations reached without convergence.")

# Interval [a, b]
a = 2
b = 3

# Find the root using Regula-Falsi method
root = regula_falsi_method(a, b)
print(f"The approximate root of the equation x³ - 2x - 5 = 0 in the interval [{a}, {b}] is: {root:.6f}")
```

```python
#Q.3b.2  Write Python program to estimate the value of the integral 2 to 4 (2x²-4x+1)dx using Trapezoidal rule (n=5).

def f(x):
    return 2 * x**2 - 4 * x + 1

def trapezoidal_rule(a, b, n):
    h = (b - a) / n  # Calculate the width of each subinterval
    integral = (f(a) + f(b)) / 2  # Initialize with the first and last term

    # Apply the Trapezoidal rule
    for i in range(1, n):
        x = a + i * h
        integral += f(x)  # Sum the middle terms

    integral *= h  # Multiply by the width of the subintervals
    return integral

# Define the interval [a, b] and number of subintervals n
a = 2
b = 4
n = 5  # Number of subintervals

# Calculate the integral using the Trapezoidal rule
result = trapezoidal_rule(a, b, n)
print(f"The estimated value of the integral is: {result:.6f}")
```
