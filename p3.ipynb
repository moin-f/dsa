```python
# Q.3 a. 1.Write Python program to estimate the value of the integration sin(x)dx of limits 0 to pi using Simpson’s (1/3)rd rule (n=6).
import numpy as np

def f(x):
    """Function to integrate."""
    return np.sin(x)

def simpsons_rule(a, b, n):
    """Estimate the integral of f(x) from a to b using Simpson's (1/3) rule."""
    if n % 2 != 0:
        raise ValueError("n must be an even number.")
    
    h = (b - a) / n  # Step size
    integral = f(a) + f(b)  # f(a) + f(b)
    
    # Apply Simpson's rule
    for i in range(1, n, 2):
        integral += 4 * f(a + i * h)  # Odd indices
    
    for i in range(2, n-1, 2):
        integral += 2 * f(a + i * h)  # Even indices
    
    integral *= h / 3  # Final multiplication by h/3
    return integral

# Define the interval and number of subintervals
a = 0
b = np.pi
n = 6

# Estimate the integral
integral_value = simpsons_rule(a, b, n)

print(f"The estimated value of the integral ∫ sin(x) dx from 0 to π is: {integral_value:.6f}")
```

```python
#Q.3a.2 write python program to evaluate third order forward difference of the given data 
# x  0 1 2 3    y=f(x) 1 0 1 10
import numpy as np

# Given data
X = np.array([0, 1, 2, 3])
Y = np.array([1, 0, 1, 10])

# Calculate forward differences
def forward_difference(Y):
    n = len(Y)
    # Create a table for storing differences
    diff_table = np.zeros((n, n))
    diff_table[:, 0] = Y  # First column is original Y values

    # Calculate forward differences
    for j in range(1, n):
        for i in range(n - j):
            diff_table[i][j] = diff_table[i + 1][j - 1] - diff_table[i][j - 1]

    return diff_table

# Compute the forward difference table
diff_table = forward_difference(Y)

# Output the forward difference table
print("Forward Difference Table:")
for i in range(len(diff_table)):
    print(" | ".join(f"{diff_table[i][j]:.4f}" for j in range(len(diff_table[i])) if diff_table[i][j] != 0))
```

```python
#Q.3b.1 write python program to evaluate f93.5of the given data 
# x 1 2 3 4 5  y=f(x) 30 50 55 40 11
import numpy as np

# Given data points
X = np.array([1, 2, 3, 4, 5])
Y = np.array([30, 50, 55, 40, 11])

def linear_interpolation(x, x0, x1, y0, y1):
    """Perform linear interpolation."""
    return y0 + (y1 - y0) * (x - x0) / (x1 - x0)

def estimate_f(target_x):
    """Estimate the value of f(target_x) using linear interpolation."""
    # Check if target_x is within the range of X
    if target_x < X[0] or target_x > X[-1]:
        raise ValueError("Target x is out of bounds of the provided data.")

    # Find the interval for target_x
    for i in range(len(X) - 1):
        if X[i] <= target_x <= X[i + 1]:
            # Perform linear interpolation
            return linear_interpolation(target_x, X[i], X[i + 1], Y[i], Y[i + 1])
    
    return None  # Just in case we don't find a suitable interval

# Estimate f(3.5)
target_x = 3.5
estimated_value = estimate_f(target_x)

print(f"The estimated value of f({target_x}) is: {estimated_value:.4f}")
```

```python
#Q.3b.2 write python program to estimate the value of the integral 2 to10 (1/(1+x))dx usiong traprzoidal rule (n=5)
import numpy as np

# Define the function to integrate
def f(x):
    return 1/(1 + x)

# Trapezoidal Rule implementation
def trapezoidal_rule(a, b, n):
    h = (b - a) / n  # Step size
    integral_value = 0.5 * (f(a) + f(b))  # Start with the endpoints

    # Calculate the area for each trapezoid
    for i in range(1, n):
        integral_value += f(a + i * h)

    integral_value *= h  # Final result
    return integral_value

# Set the limits of integration and the number of intervals
a = 2
b = 10
n = 5

# Estimate the integral
integral_result = trapezoidal_rule(a, b, n)
print(f"The estimated value of the integral from {a} to {b} for 1/(1 + x)dx using Trapezoidal rule is: {integral_result:.4f}")
```
